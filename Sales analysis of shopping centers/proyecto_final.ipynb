# Assignment Final
Los datos que se incluyen en el fichero `datos_ventas_centros_comerciales.csv` representan casi 100.000 compra-ventas realizadas en un grupo de Centros Comerciales durantes los años 2021 y 2022. 

Como analista del holding propietario de los centros comerciales realiza un analisis respondiendo a las siguiente preguntas.

```python
import pandas as pd

df_vv_cc = pd.read_csv('datos_ventas_centros_comerciales.csv')

df_vv_cc["invoice_date"] = pd.to_datetime(df_vv_cc.invoice_date, format="%d-%m-%Y")
df_vv_cc["day_name"] = df_vv_cc["invoice_date"].apply(lambda x: x.day_name())

filtro_fechas = ((df_vv_cc.invoice_date <= "2022-12-31") & (df_vv_cc.invoice_date >= "2021-01-01"))
df_vv_cc_fil = df_vv_cc[filtro_fechas]
```

# 1. ¿Cuál es el total de ingresos generados por todas las ventas registradas?
Guarda el resultado en una variable llamada `total_ingresos`



```python
total_ingresos = (df_vv_cc_fil["quantity"] * df_vv_cc_fil["price"]).sum()
print("El total de ingresos es:", total_ingresos)
```

# 2. ¿Cuál es el centro comercial que ha generado más ventas?
Guarda el resultado en una variable llamada `centro_comercial_mas_ventas`


```python
df_vv_cc_fil["total_sales"] = df_vv_cc_fil["quantity"] * df_vv_cc_fil["price"]
ventas_cc = df_vv_cc_fil.groupby("shopping_mall")["total_sales"].sum()

centro_comercial_mas_ventas = ventas_cc.idxmax()
print("El centro comercial que ha generado mas ventas es:", centro_comercial_mas_ventas)
```

# 3. ¿Cuál es la categoría de producto que ha generado mas ventas? 
Guarda el resultado en una variable llamada `categoria_mas_vendida`


```python
ventas_categoria = df_vv_cc_fil.groupby("category")["total_sales"].sum()
categoria_mas_vendida = ventas_categoria.idxmax()
print("La categoria mas vendida es:", categoria_mas_vendida)
```

# 4. ¿Cuál es el producto más caro vendido?
Guarda el resultado en una variable llamada `categoria_producto_mas_caro`



```python
producto_mas_caro = df_vv_cc_fil.loc[df_vv_cc_fil["price"].idxmax()]
categoria_producto_mas_caro = producto_mas_caro["category"]
print("El producto más caro vendido es:", categoria_producto_mas_caro)
```

# 5. ¿Cuál es la factura más antigua registrada en el dataset?
Guarda el resultado en una variable llamada `factura_mas_antigua`


```python
factura_mas_antigua = df_vv_cc_fil.loc[df_vv_cc_fil["invoice_date"].idxmin()]
print("La factura más antigua del dataset es:")
print()
print(factura_mas_antigua)
```

# 6. ¿Cuál es la cantidad promedio de productos vendidos por transacción?
Guarda el resultado en una variable llamada `cantidad_promedio_por_transaccion`



```python
cantidad_promedio_por_transaccion = df_vv_cc_fil["quantity"].mean()
print("La cantidad promedio de productos vendidos por transacción es", cantidad_promedio_por_transaccion)
```

# 7. ¿Cuál es el día con más ventas registradas?
Guarda el resultado en una variable llamada `dia_con_mas_ventas`


```python
dia_con_mas_ventas = df_vv_cc_fil["day_name"].value_counts().idxmax()
print("El dia con más ventas registradas es:", dia_con_mas_ventas)
```

# 8. ¿Cuál es el cliente que más ha gastado en total?
Guarda el resultado en una variable llamada `cliente_que_mas_gasto`


```python
# Group by customer_id and sum the total_spent
total_gastado_por_cliente = df_vv_cc_fil.groupby("customer_id")["total_sales"].sum()

# Find the customer_id with the maximum total_spent
cliente_que_mas_gasto = total_gastado_por_cliente.idxmax()
print("El cliente que mas ha gastado en total es:", cliente_que_mas_gasto)
```

# 9. ¿Cuál es la factura con el precio total más alto?
Guarda el resultado en una variable llamada `factura_max_valor`


```python
factura_max_valor = df_vv_cc_fil.loc[df_vv_cc_fil["total_sales"].idxmax()]
print("La factura con el precio más alto es:")
print()
print(factura_max_valor)
```

# 10. ¿Cuál es la distribución porcentual de ventas por categoría de producto?
Guarda el resultado en una variable llamada `distribucion_porcentual_ventas_por_categoria`. El tipo de `distribucion_porcentual_ventas_por_categoria` debe ser `pandas.core.series.Series` donde las etiquetas deben ser las categorias y los valores deben ser la distribucion porcentual de las ventas por categoría


```python
total_ventas_por_categoria = df_vv_cc_fil.groupby("category")["total_sales"].sum()
distribucion_porcentual_ventas_por_categoria = (total_ventas_por_categoria / total_ventas_por_categoria.sum()) * 100
print("La distribución porcentual de ventas por categorias de producto es:")
print()
print(distribucion_porcentual_ventas_por_categoria)
print(type(distribucion_porcentual_ventas_por_categoria))
```

# 11. ¿Cuál es el día de la semana con más ventas registradas en cada centro comercial?
Guarda el resultado en una variable llamada `dia_semana_mas_ventas_por_centro`. El tipo de `dia_semana_mas_ventas_por_centro` debe ser `pandas.core.series.Series` donde las etiquetas deben ser el nombre de los centros comerciales y los valores deben ser el nombre del dia de la semana con mas ventas.

```python
dia_semana_mas_ventas_por_centro = df_vv_cc_fil.groupby("shopping_mall")["day_name"].agg(lambda x: x.value_counts().idxmax())

print("El día de la smena con más ventas registradas en cada centro comercial es:")
print()
print(dia_semana_mas_ventas_por_centro)
print(type(dia_semana_mas_ventas_por_centro))
```
